```python
import os
import requests
from dotenv import load_dotenv
from tabulate import tabulate

load_dotenv()


def get_company_data(symbol):
    api_key = os.getenv('ALPHA_VANTAGE_API_KEY')
    url = f'https://www.alphavantage.co/query?function=OVERVIEW&symbol={symbol}&apikey=UKCUS1RPG2ABJIL8'
    response = requests.get(url)
    data = response.json()
    return data

if __name__ == "__main__":
    symbol = input("Enter the stock symbol: ").upper()  # Get the stock symbol as input and convert it to uppercase
    full_name = symbol_names.get(symbol, "N/A")  # Get the full name based on the symbol entered by the user
    print(f"Full Name: {full_name}")

    data = get_company_data(symbol)

    # Data processing for tabulate
    table = []
    for key, value in data.items():
        table.append([key, value])

    # Printing the tabulated data
    headers = ["Attribute", "Value"]
    print(tabulate(table, headers, tablefmt="grid"))

```
